```python
from pattern import color, convert2markers
def convert2markers(coordinatenliste):
    markers=[]
    for row in coordinatenliste:
        for stitch in row:
           markers.append(
                CircleMarker(
                    location=stitch,
                    radius=2,
                    color="white",
                    fill_color=str(color(loader.lookup(*stitch))),
                    fill_opacity=0.9,
                    weight=0
                ))

    #return markers
#print(convert2markers(coordinatesf(20,20,0,10)))
```
