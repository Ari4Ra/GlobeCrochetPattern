```python
def convert2features(coordinatenliste):
    Features=[]
    for row in coordinatenliste:
        for stitch in row:
            Features.append({
            'type': 'Feature',
            'geometry': {
                'type': 'Point',
                'coordinates': reversed(stitch),  # Longitude, Latitude for New York City
            }})
    return Features
#convert2features(coordinatesf(3,3,0,30))
```

```python
from ipyopenlayers import Map, GeoJSON, RasterTileLayer
m = Map(
    center=[0, 0],
    zoom=2
)
# Add layer
layer=RasterTileLayer()
m.add_layer(layer)

geojson_data = {
    'type': 'FeatureCollection',
    'features': convert2features(coordinatesf(15,15,0,30))}
geojson_style = {
    'Point': {
        'radius': 8,
        'fillColor': '#ff7800',
        'color': '#000',
        'weight': 2,
        'opacity': 1,
        'fillOpacity': 0.8,
    },
    'LineString': {
        'color': '#00f',
        'weight': 3,
        'opacity': 0.8,
    },
    'Polygon': {
        'strokeColor': '#00f',
        'strokeWidth': 2,
        'strokeOpacity': 0.8,
        'fillColor': '#ff7800',
        'fillOpacity': 0.5,
    },
    'MultiPolygon': {
        'strokeColor': '#f00',
        'strokeWidth': 2,
        'strokeOpacity': 0.8,
        'fillColor': '#0f0',
        'fillOpacity': 0.4,
    },
}

# Create and add a GeoJSON layer to the map
widget = GeoJSON(
    data=geojson_data,
    style=geojson_style,
    visible=True
)
m.add_layer(widget)
# Display the map
m
```

```python
import rioxarray
from pyproj import Transformer

def lookup(lat, lon):
    transformer = Transformer.from_crs("EPSG:4326", xds.rio.crs, always_xy=True)
    xx, yy = transformer.transform(lon, lat)

    return xds.sel(x=xx, y=yy, method="nearest").values
```
