```python
from pattern import StitchCoordinates, color
from ipyleaflet import Map, GeoJSON, CircleMarker, LayerGroup

stitch_coordinates=StitchCoordinates(4,3.5,1,5)

def convert2markers(coordinatenliste, lloader):
    markers=[]
    col=lloader.lookup_majority_batch_nested(coordinatenliste,5,5,5)
    for i,row in enumerate(coordinatenliste):
        for j,stitch in enumerate(row):
            markers.append(
                CircleMarker(
                    location=stitch,
                    radius=2,
                    color="white",
                    fill_color=str(color(col[i][j])),
                    fill_opacity=0.9,
                    weight=0
                ))

    return markers


m = Map(
    center=[0, 0],
    zoom=1
)

m.add(LayerGroup(layers=convert2markers(stitch_coordinates.coordinates(),loader)))

m
```
